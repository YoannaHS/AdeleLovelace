# 4. Productos más y menos vendidos

```python
import pandas as pd
import matplotlib.pyplot as plt

# Combinar los DataFrames de las tiendas en uno solo llamado 'all_stores'
all_stores = pd.concat([tienda_1, tienda_2, tienda_3, tienda_4], keys=['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'], names=['Tienda'])

# Agrupar por tienda y producto, contando las ocurrencias de cada producto
ventas_por_producto = all_stores.groupby(['Tienda', 'Producto'])['Producto'].count().reset_index(name='Cantidad Vendida')

# Mostrar el DataFrame resultante
print(ventas_por_producto)
```

```python
import pandas as pd

# ... (código anterior para crear 'ventas_por_producto') ...

def productos_top_bottom(tienda, top_n=5):
    """
    Encuentra los productos más y menos vendidos para una tienda específica.

    Args:
        tienda (str): Nombre de la tienda (e.g., 'Tienda 1').
        top_n (int): Número de productos top/bottom a mostrar (default: 5).

    Returns:
        tuple: Dos DataFrames, uno con los productos más vendidos y otro con los menos vendidos.
    """
    ventas_tienda = ventas_por_producto[ventas_por_producto['Tienda'] == tienda]
    top_vendidos = ventas_tienda.sort_values(by='Cantidad Vendida', ascending=False).head(top_n)
    menos_vendidos = ventas_tienda.sort_values(by='Cantidad Vendida', ascending=True).head(top_n)
    return top_vendidos, menos_vendidos

tiendas = ventas_por_producto['Tienda'].unique()  # Obtener la lista de tiendas únicas

for tienda in tiendas:
    top_vendidos, menos_vendidos = productos_top_bottom(tienda)
    print(f"\n--- Resultados para {tienda} ---")
    print("\nTop 5 productos:\n", top_vendidos)
    print("\nBottom 5 productos:\n", menos_vendidos)
```

```python
import matplotlib.pyplot as plt
import numpy as np

def plot_top_bottom_products(tienda, top_n=5):
    """
    Genera un gráfico de barras para los productos más y menos vendidos de una tienda.

    Args:
        tienda (str): Nombre de la tienda (e.g., 'Tienda 1').
        top_n (int): Número de productos top/bottom a mostrar (default: 5).
    """

    top_vendidos, menos_vendidos = productos_top_bottom(tienda, top_n)

    # Combinar los DataFrames de top y bottom en uno solo
    top_vendidos['Ranking'] = 'Top ' + top_vendidos['Cantidad Vendida'].astype(str)
    menos_vendidos['Ranking'] = 'Bottom ' + menos_vendidos['Cantidad Vendida'].astype(str)

    all_products = pd.concat([top_vendidos, menos_vendidos])

    # Crear el gráfico de barras
    plt.figure(figsize=(12, 6))
    plt.barh(all_products['Producto'], all_products['Cantidad Vendida'], color=np.where(all_products['Ranking'].str.startswith('Top'), 'green', 'red'))

    plt.title(f'Top {top_n} y Bottom {top_n} Productos Vendidos en {tienda}')
    plt.xlabel('Cantidad Vendida')
    plt.ylabel('Producto')

    # Agregar etiquetas de cantidad vendida a las barras
    for index, value in enumerate(all_products['Cantidad Vendida']):
        plt.text(value, index, str(value), color='black', ha='left', va='center')

    plt.gca().invert_yaxis()  # Invertir el eje Y para que los productos más vendidos estén en la parte superior
    plt.tight_layout()
    plt.show()

# Llamar a la función para cada tienda
for tienda in tiendas:
    plot_top_bottom_products(tienda)
```
